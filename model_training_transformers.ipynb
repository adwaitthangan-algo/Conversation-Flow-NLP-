```python
import torch
from torch.utils.data import DataLoader, TensorDataset
from transformers import BertTokenizer, AdamW

# Load BERT tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Tokenize and pad the text data
max_len = 128  # You can adjust this based on your requirements
X_texts = [x[0] for x in X]
X_labels = [x[1] for x in X]
encoded_data = tokenizer(X_texts, padding=True, truncation=True, max_length=max_len, return_tensors="pt")

input_ids = encoded_data["input_ids"]
attention_mask = encoded_data["attention_mask"]

# Create DataLoader
dataset = TensorDataset(input_ids, attention_mask, torch.tensor(X_labels))
batch_size = 4  # You can adjust this as needed
data_loader = DataLoader(dataset, batch_size=batch_size)

# Define loss function, optimizer, and model
loss_fn = torch.nn.CrossEntropyLoss()
optimizer = AdamW(model.parameters(), lr=1e-5)  # You can adjust the learning rate

# Training loop
num_epochs = 14  # You can adjust this
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for batch in data_loader:
        input_ids_batch, attention_mask_batch, labels_batch = batch
        optimizer.zero_grad()
        outputs = model(input_ids_batch, attention_mask=attention_mask_batch, labels=labels_batch)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(data_loader)
    print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {avg_loss:.4f}")
```

```python

import torch
from torch.utils.data import DataLoader, TensorDataset
from transformers import BertTokenizer, AdamW, BertForSequenceClassification

# Check GPU availability
if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"Using GPU: {torch.cuda.get_device_name(0)}")
else:
    device = torch.device("cpu")
    print("No GPU available, using CPU instead.")

# Load BERT tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Tokenize and pad the text data
max_len = 80  # You can adjust this based on your requirements
X_texts = [x[0] for x in X]
X_labels = [x[1] for x in X]
encoded_data = tokenizer(X_texts, padding=True, truncation=True, max_length=max_len, return_tensors="pt")

input_ids = encoded_data["input_ids"]
attention_mask = encoded_data["attention_mask"]

# Move data to GPU
input_ids = input_ids.to(device)
attention_mask = attention_mask.to(device)
X_labels = torch.tensor(X_labels).to(device)

# Create DataLoader
dataset = TensorDataset(input_ids, attention_mask, X_labels)
batch_size = 8  # You can adjust this as needed
data_loader = DataLoader(dataset, batch_size=batch_size)

# Load BERT model
model = BertForSequenceClassification.from_pretrained("bert-base-uncased",
                                                      num_labels=32,
                                                      output_hidden_states=False)

# Move model to GPU
model.to(device)

# Define loss function, optimizer, and model
loss_fn = torch.nn.CrossEntropyLoss()
optimizer = AdamW(model.parameters(), lr=1e-5)  # You can adjust the learning rate

# Training loop
num_epochs = 24  # You can adjust this
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for batch in data_loader:
        input_ids_batch, attention_mask_batch, labels_batch = batch
        optimizer.zero_grad()
        outputs = model(input_ids_batch, attention_mask=attention_mask_batch, labels=labels_batch)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(data_loader)
    print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {avg_loss:.4f}")




# import torch
# from torch.utils.data import DataLoader, TensorDataset
# from transformers import BertTokenizer, AdamW

# # Load BERT tokenizer
# tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# # Tokenize and pad the text data
# max_len = 64  # You can adjust this based on your requirements
# X_texts = [x[0] for x in X]
# X_labels = [x[1] for x in X]
# encoded_data = tokenizer(X_texts, padding=True, truncation=True, max_length=max_len, return_tensors="pt")

# input_ids = encoded_data["input_ids"]
# attention_mask = encoded_data["attention_mask"]

# # Create DataLoader
# dataset = TensorDataset(input_ids, attention_mask, torch.tensor(X_labels))
# batch_size = 6  # You can adjust this as needed
# data_loader = DataLoader(dataset, batch_size=batch_size)

# # Define loss function, optimizer, and model
# loss_fn = torch.nn.CrossEntropyLoss()
# optimizer = AdamW(model.parameters(), lr=1e-4)  # You can adjust the learning rate

# # Training loop
# num_epochs = 18  # You can adjust this
# for epoch in range(num_epochs):
#     model.train()
#     total_loss = 0
#     for batch in data_loader:
#         input_ids_batch, attention_mask_batch, labels_batch = batch
#         optimizer.zero_grad()
#         outputs = model(input_ids_batch, attention_mask=attention_mask_batch, labels=labels_batch)
#         loss = outputs.loss
#         loss.backward()
#         optimizer.step()
#         total_loss += loss.item()
#     avg_loss = total_loss / len(data_loader)
#     print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {avg_loss:.4f}")
```
